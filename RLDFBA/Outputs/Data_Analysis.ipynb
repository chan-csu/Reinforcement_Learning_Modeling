```python
import pickle
import pandas as pd
import plotly.figure_factory as ff
import plotly
```

```python
with open('Ecoli_1_0.pkl','rb') as M:
    M1=pickle.load(M)
with open('Ecoli_2_0.pkl','rb') as M:
    M2=pickle.load(M)
```

```python
import plotly.express as px
import plotly.graph_objects as go
```

```python
Results_Dict={'Glucose':[],
         'Starch':[],
         'Amylase':[],
         'AP1':[],
         'AP2':[],
        }
for item in M1.keys():
    Results_Dict['Glucose'].append(item[0])
    Results_Dict['Starch'].append(item[1])
    Results_Dict['Amylase'].append(item[2])
    Results_Dict['AP1'].append(M1[item])
    Results_Dict['AP2'].append(M2[item])   
    
```

```python
DF=pd.DataFrame(Results_Dict)
```

```python
fig = plotly.subplots.make_subplots(rows=1, cols=2,
                                   specs=[[{"type": "ternary"}, {"type": "ternary"}]],horizontal_spacing=0.15,
                                   column_titles=["E.coli 1","E.coli 2"])
```

```python
fig.append_trace(
    go.Scatterternary({
    'mode': 'markers',
    'a': DF['Glucose'],
    'b': DF['Starch'],
    'c': DF['Amylase'],
    'marker': {
        'symbol': 100,
        'color': DF['AP1'],
        'size': 4,
        'coloraxis':"coloraxis1",
        'line': { 'width': 2 }
    }
    ,
    "marker_symbol":"circle-open-dot",
        
}),1,1
)
fig.append_trace(
    go.Scatterternary({
    'mode': 'markers',
    'a': DF['Glucose'],
    'b': DF['Starch'],
    'c': DF['Amylase'],
    'marker': {
        'symbol': 100,
        'color': DF['AP2'],
        'coloraxis':"coloraxis1",
        'size': 4,
        'line': { 'width': 2 }
    },
    "marker_symbol":"circle-open-dot"
        
}),1,2
)
fig.update_ternaries(aaxis_title='Glucose',
                    baxis_title='Starch',
                    caxis_title="Amylase")
fig.update_layout(showlegend=False)
fig.update_coloraxes(colorbar_xpad=50)
fig.show()
```

```python
fig.write_html("First_Trial.html")
```
